```python
import pandas as pd

df=pd.read_csv('D:/Tyler_Data/Other Skills/Datascience/MMDT_GitHub/MMDT_2025_MLAI101/Htet-Aung-Phyo/NB_Classifier/fraud.csv', index_col = 0)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
df.head(5)
```

```python
y = df['Class'].values
df = df.iloc[:,1:]
X = df.drop(columns = 'Class').values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, 
                                    test_size = 0.40, 
                                    random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
ypred_test = gnb.predict(X_test)
gnb_mat = confusion_matrix(y_test, ypred_test)
gnb_report = classification_report(y_test, ypred_test)

print(gnb_mat)
print(gnb_report)


auc = roc_auc_score(y_test, ypred_test)
print(auc)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()  # scale features to [0,1]
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB(alpha=1.0)
mnb.fit(X_train_scaled, y_train)
```

```python
ypred_test = mnb.predict(X_test_scaled)
mnb_mat = confusion_matrix(y_test, ypred_test)
mnb_report = classification_report(y_test, ypred_test)

print(mnb_mat)
print(mnb_report)


auc = roc_auc_score(y_test, ypred_test)
print(auc)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB(alpha=1.0)
bnb.fit(X_train, y_train)
```

```python
ypred_test = bnb.predict(X_test)
bnb_mat = confusion_matrix(y_test, ypred_test)
bnb_report = classification_report(y_test, ypred_test)

print(bnb_mat)
print(bnb_report)


auc = roc_auc_score(y_test, ypred_test)
print(auc)
```
